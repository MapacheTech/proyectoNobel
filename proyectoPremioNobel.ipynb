```python
top_country = nobel['birth_country'].value_counts().idxmax()

top_country
```

```python
# Recuento de premios por país de nacimiento, de mayor a menor
conteo_paises = nobel['birth_country'].value_counts()

# Ver todos los países (sin truncar la salida)
pd.set_option('display.max_rows', None)
print(conteo_paises)
pd.reset_option('display.max_rows')

# Top 10 países
top10_paises = conteo_paises.head(10)
print(top10_paises)
```

```python
# Gráfico del top 10 de países, ordenado de mayor a menor
grafica_top10 = sns.barplot(x=top10_paises.values, y=top10_paises.index, order=top10_paises.index)

grafica_top10.set_title('Top 10 Países por Número de Premios Nobel (Nacimiento)')
grafica_top10.set_xlabel('Número de Premios')
grafica_top10.set_ylabel('País de Nacimiento')
plt.show()
```
